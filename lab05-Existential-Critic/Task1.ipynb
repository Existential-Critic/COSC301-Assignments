
### Task 1.1: Join/Merge three datasets together

Your task is to take three separate datasets (`grades1a.csv`, `grades1b.csv`, `grades1c.csv` -> find them in the **data** folder) and merge them together.

````python
#Your Answer Here

import pandas as pd

#Read the CSV files and save them to dataframes
g1a = pd.read_csv('data/grades1a.csv')
g1b = pd.read_csv('data/grades1b.csv')
g1c = pd.read_csv('data/grades1c.csv')
#Save the three dataframes into a list to be iterated through
csvFiles = [g1a,g1b,g1c]

#Create an empty list to contain the organised student IDs
studentIDList = []
#Iterate through each of the dataframes and save each unique student ID
for file in csvFiles:
    for id in file['Student ID']:
        if id not in studentIDList:
            studentIDList.append(id)
#Sort the student IDs in numerical order
studentIDList.sort()

#Create a new list of headers for a new dataframe, starting with Student ID
headerList = ['Student ID']
#Iterate through each dataframe and save each unique header
for file in csvFiles:
    for item in file.columns.to_list():
        if item not in headerList:
            headerList.append(item)

#Create a new dataframe which will start as an empty list and end as a list of lists
g1Merged = []
#Iterate through each unique student ID
for i in studentIDList:
    #Create an empty list for checking which information has already been saved
    usedKeys = ['Student ID']
    #Create a list to be appended to the merged list
    studentInfo = [i]
    #Iterate through each dataframe and save the data for the unique student ID
    for file in csvFiles:
        #Iterate through each row of the file
        for num in range(file.shape[0]):
            #Check if the student ID of the row is the correct ID
            if file.loc[num,'Student ID'] == i:
                #Go through each header/key and check if it has already been used for this student
                for header in headerList:
                    if header not in usedKeys:
                        #If the header has not been used, save that info to the student info
                        try:
                            studentInfo.append(file.loc[num,header])
                            usedKeys.append(header)
                        except KeyError:
                            pass
    #Add the student info to the overall list
    g1Merged.append(studentInfo)

#Turn the list into a dataframe
g1MergedDF = pd.DataFrame(g1Merged,columns=headerList)
#Save to a new csv file
g1MergedDF.to_csv('Task 1/task1-1.csv')
````

````python
#Your Answer Here

import pandas as pd

#Read the CSV files and save them to dataframes
g2a = pd.read_csv('data/grades2a.csv')
g2b = pd.read_csv('data/grades2b.csv')
#Save the two dataframes into a list to be iterated through
csvFiles = [g2a,g2b]

#Create an empty list to contain the organised student IDs
studentIDList = []
#Iterate through each of the dataframes and save each unique student ID
for file in csvFiles:
    if 'Student ID' in file.columns:
        for id in file['Student ID']:
            if id not in studentIDList:
                studentIDList.append(id)
    else:
        for id in file['SID']:
                if id not in studentIDList:
                    studentIDList.append(id)
# #Sort the student IDs in numerical order
studentIDList.sort()

#Create a new list of headers for a new dataframe, starting with Student ID
headerList = ['Student ID']
#Iterate through each dataframe and save each unique header
for file in csvFiles:
    for item in file.columns.to_list():
        if item not in headerList:
            if item != 'SID':
                headerList.append(item)

#Create a new dataframe which will start as an empty list and end as a list of lists
g2Merged = []
#Iterate through each unique student ID
for i in studentIDList:
    #Create an empty list for checking which information has already been saved
    usedKeys = ['Student ID']
    #Create a list to be appended to the merged list
    studentInfo = [i]
    #Iterate through each dataframe and save the data for the unique student ID
    for file in csvFiles:
        #Iterate through each row of the file
        for num in range(file.shape[0]):
            #Check if the student ID of the row is the correct ID
            if 'Student ID' in file.columns:
                if file.loc[num,'Student ID'] == i:
                    #Go through each header/key and check if it has already been used for this student
                    for header in headerList:
                        if header not in usedKeys:
                            #If the header has not been used, save that info to the student info
                            try:
                                studentInfo.append(file.loc[num,header])
                                usedKeys.append(header)
                            except KeyError:
                                pass
            else:
                if file.loc[num,'SID'] == i:
                    #Go through each header/key and check if it has already been used for this student
                    for header in headerList:
                        if header not in usedKeys:
                            #If the header has not been used, save that info to the student info
                            try:
                                studentInfo.append(file.loc[num,header])
                                usedKeys.append(header)
                            except KeyError:
                                pass
    #Add the student info to the overall list
    g2Merged.append(studentInfo)

#Turn the list into a dataframe
g2MergedDF = pd.DataFrame(g2Merged,columns=headerList)
#Save to a new csv file
g2MergedDF.to_csv('Task 1/task1-2.csv')
````

### Task 1.3: Merge dataframes and keep only overlapping rows

Often it is useful to try a bunch of different merge operations on a dataset to identify what parts of the data overlap, and what parts are disparate.
With the overlapping rows, you can then look at the value from the left, and the value from the right to compare them to see if there is any ambiguity if you chose to merge the dataframes.

In this question, you will use the same data as in Task 2 (`grades2a.csv` and `grades2b.csv`) and try to identify which rows are duplicates, or appear in both datasets.

````python
#Your Answer Here

import pandas as pd

#Read the CSV files and save them to dataframes
g2a = pd.read_csv('data/grades2a.csv')
g2b = pd.read_csv('data/grades2b.csv')

#Create two empty lists to contain the respective organised student IDs
studentIDList = []
SIDList = []
#Iterate through each of the dataframes and save each unique student ID and unique SID
for id in g2a['Student ID']:
    if id not in studentIDList:
        studentIDList.append(id)
for id in g2b['SID']:
    if id not in SIDList:
        SIDList.append(id)
# #Sort the student IDs in numerical order
studentIDList.sort()
SIDList.sort()
#Find the intersection of the two lists
studentIDSIDList = [item for item in studentIDList if item in SIDList]

#Make two lists of respective headers
g2aHeaders = list(g2a.columns)
g2bHeaders = list(g2b.columns)

#Create a new dataframe which will start as an empty list and end as a list of lists
g2Merged = []
#Iterate through each unique student ID
for i in studentIDSIDList:
    #Create a list to be appended to the merged list
    studentInfo = []
    #Create an empty list for checking which information has already been saved
    usedG2aKeys = []
    usedG2bKeys = []
    #Iterate through each dataframe and save the data for the unique student ID
    for num in range(g2a.shape[0]):
        #Check if the student ID of the row is the correct ID
        if g2a.loc[num,'Student ID'] == i:
            #Go through each header/key and check if it has already been used for this student
            for header in g2aHeaders:
                if header not in usedG2aKeys:
                    #If the header has not been used, save that info to the student info
                    try:
                        studentInfo.append(g2a.loc[num,header])
                        usedG2aKeys.append(header)
                    except KeyError:
                        pass
    for num in range(g2b.shape[0]):
        #Check if the student ID of the row is the correct ID
        if g2b.loc[num,'SID'] == i:
            #Go through each header/key and check if it has already been used for this student
            for header in g2bHeaders:
                if header not in usedG2bKeys:
                    #If the header has not been used, save that info to the student info
                    try:
                        studentInfo.append(g2b.loc[num,header])
                        usedG2bKeys.append(header)
                    except KeyError:
                        pass
    g2Merged.append(studentInfo)

#Create the headers for the new CSV file
newHeaders = ['Name_x','Student ID','Chemistry_x','Physics_x','Math_x','Name_y','SID','Chemistry_y','Physics_y','Math_y']

#Turn the merged list into a dataframe
g2MergedDF = pd.DataFrame(g2Merged,columns=newHeaders)
#Save to a new csv file
g2MergedDF.to_csv('Task 1/task1-3.csv')
````
